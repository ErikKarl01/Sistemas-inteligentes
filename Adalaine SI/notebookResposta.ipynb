## Separar target:

```python
x = dados.iloc[:, :-1]
y = dados.iloc[:, -1]

y.head(n=10)
```

```python
import numpy as np

y = np.array(y)

y = np.where(y == 'g', 1, -1)
y
```

## Normalização de atributos:

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x = scaler.fit_transform(x)

print(x.mean(axis=0))
print(x.std(axis=0))
```

## Remover coluna com valores constantes:

```python
from sklearn.feature_selection import VarianceThreshold

selector = VarianceThreshold()

x = selector.fit_transform(x)

print(x.mean(axis=0))
print(x.std(axis=0))
```

## Separação de treino e teste:

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

## Treinar modelos baseados na taxa de aprendizado:

```python
from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

N = 200
taxas = [0.1, 0.01, 0.001, 0.0004, 0.00001, 0.000001, 0.00000001]

for eta in taxas:

    modelo = SGDClassifier(
        loss="squared_error",
        learning_rate="constant",
        eta0=eta,
        max_iter=1,
        tol=None,
        random_state=42
    )

    perdas = []

    for epoca in range(N):

        modelo.partial_fit(X_train, Y_train, classes=np.unique(Y_train))

        previsoes = modelo.predict(X_train)

        mse = mean_squared_error(Y_train, previsoes)

        perdas.append(mse)

    plt.figure(figsize=(8, 5))

    plt.plot(
        perdas,
        label=f"Taxa de aprendizagem (η) = {eta}"
    )

    plt.xlabel("Épocas de treinamento")
    plt.ylabel("Erro Quadrático Médio (MSE)")

    plt.title(f"Curva de Convergência - η = {eta}")

    plt.ylim(min(perdas) - 0.01, max(perdas) + 0.01)

    plt.yticks(
        np.linspace(
            min(perdas),
            max(perdas),
            15
        )
    )

    plt.legend()

    plt.grid(True)

    plt.show()
```

## Treino com o melhor taxa:

```python

modelo = SGDClassifier(
        loss="squared_error",
        learning_rate="constant",
        eta0=0.0004,
        max_iter=1,
        tol=None,
        random_state=42
    )

for epoca in range(N):

    modelo.partial_fit(X_train, Y_train, classes=np.unique(Y_train))

    previsoes = modelo.predict(X_train)

    mse = mean_squared_error(Y_train, previsoes)
```

## Métricas:

```python
from sklearn.metrics import (
    accuracy_score,
    f1_score,
    confusion_matrix,
    precision_score,
    recall_score
)

# previsões no conjunto de teste
Y_pred = modelo.predict(X_test)

# métricas
acuracia = accuracy_score(Y_test, Y_pred)

precisao = precision_score(Y_test, Y_pred)

recall = recall_score(Y_test, Y_pred)

f1 = f1_score(Y_test, Y_pred)

matriz = confusion_matrix(Y_test, Y_pred)

# resultados
print(f"Acurácia: {acuracia:.4f}")

print(f"Precisão: {precisao:.4f}")

print(f"Recall: {recall:.4f}")

print(f"F1-Score: {f1:.4f}")

print("\nMatriz de Confusão:")

print(matriz)
```
